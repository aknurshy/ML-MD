```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import imutils
import zipfile
```

```python
#Loading of image
img = cv2.imread('/Users/aknursh/Desktop/MASTER/ML/original.jpg', -1)
plt.imshow(img)
```

```python
#Resizing image, after getting the original size
height, width = img.shape[:2]
print('height =', height)
print('width =', width)
img_resize = cv2.resize(img, (55, 100))
plt.imshow(img_resize)
```

```python
#Rotation of image
#Calculating the center
center = (width / 2, height / 2)

#Defining angles of rotation and scaling (no scaling)
angle90 = 90
angle180 = 180
angle270 = 270
scale = 1.0
```

```python
#Angle of 90 degree
matrix90 = cv2.getRotationMatrix2D(center, angle90, scale)
rotation90 = cv2.warpAffine(img, matrix90, (width, height))
plt.imshow(rotation90)
```

```python
#Angle of 180 degree
matrix180 = cv2.getRotationMatrix2D(center, angle180, scale)
rotation180 = cv2.warpAffine(img, matrix180, (width, height))
plt.imshow(rotation180)
```

```python
#Angle of 270 degree
matrix270 = cv2.getRotationMatrix2D(center, angle270, scale)
rotation270 = cv2.warpAffine(img, matrix270, (width, height))
plt.imshow(rotation270)
```

```python
#Transformation of image
rows, columns, ch = img.shape
pts1 = np.float32([[50,50],[200,100],[10,150]])
pts2 = np.float32([[10,10],[200,100],[0,100]])
Mat = cv2.getAffineTransform(pts1,pts2)
dst = cv2.warpAffine(img,Mat,(columns, rows))
plt.subplot(121)
plt.imshow(img)
plt.title('Before')
plt.subplot(122)
plt.imshow(dst)
plt.title('After')
plt.show()
```
